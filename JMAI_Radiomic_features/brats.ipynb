```python
from sklearn.metrics import f1_score, make_scorer
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_val_score
import math
import numpy as np

# Initialize Stratified K-Fold
kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Initialize lists to store results
cv_scores = []
confidence_scores_all = []

# Perform cross-validation
for train_index, test_index in kfold.split(X_train_scaled, y_train):
    # Split data
    X_train_fold, X_test_fold = X_train_scaled[train_index], X_train_scaled[test_index]
    y_train_fold, y_test_fold = y_train[train_index], y_train[test_index]

    # Fit the classifier on the training fold
    xgb_classifier.fit(X_train_fold, y_train_fold)

    # Predict on the test fold
    val_probabilities = xgb_classifier.predict_proba(X_test_fold)

    # Calculate confidence scores (max probability) for each prediction in the fold
    confidence_scores_fold = val_probabilities.max(axis=1)
    confidence_scores_all.extend(confidence_scores_fold)

    # Calculate F1 score for the fold and append to the list
    val_predictions = xgb_classifier.predict(X_test_fold)
    fold_f1_score = f1_score(y_test_fold, val_predictions)
    cv_scores.append(fold_f1_score)

# Calculate mean and standard error of CV scores
mean_f1_score = np.mean(cv_scores)
sem = np.std(cv_scores, ddof=1) / math.sqrt(len(cv_scores))

# Determine the margin of error (95% confidence)
z_score = 1.96
margin_of_error = z_score * sem

print(f"Mean F1-Score: {mean_f1_score:.2f}")
print(f"95% Confidence Interval: {mean_f1_score:.2f} +/- {margin_of_error:.2f}")

# Example: Display first few confidence scores
print(f"Sample Confidence Scores: {confidence_scores_all[:10]}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Load the MNIST dataset
from sklearn.datasets import fetch_openml
mnist = fetch_openml("mnist_784")
X = mnist.data.astype("float64")
y = mnist.target.astype("int64")

# Center the data
mean = np.mean(X, axis=0)
X_centered = X - mean

# Calculate the covariance matrix
cov_matrix = np.cov(X_centered, rowvar=False)

# Compute eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eigh(cov_matrix)

# Sort eigenvalues and eigenvectors in descending order
sorted_indices = np.argsort(eigenvalues)[::-1]
eigenvalues = eigenvalues[sorted_indices]
eigenvectors = eigenvectors[:, sorted_indices]

# Select a number of top eigenvalues/components
num_components = 50
top_eigenvalues = eigenvalues[:num_components]
top_eigenvectors = eigenvectors[:, :num_components]

# Project the data onto the selected components
X_pca = np.dot(X_centered, top_eigenvectors)

# Reconstruct the data using the selected components
X_reconstructed = np.dot(X_pca, top_eigenvectors.T) + mean

# Reshape the reconstructed data back to the original image shape (28x28)
reconstructed_images = X_reconstructed.reshape(-1, 28, 28)

# Plot some original and reconstructed images
n_samples = 5
plt.figure(figsize=(10, 4))

for i in range(n_samples):
    plt.subplot(2, n_samples, i + 1)
    plt.imshow(X[i].reshape(28, 28), cmap=plt.cm.gray)
    plt.title("Original")
    plt.axis("off")

    plt.subplot(2, n_samples, n_samples + i + 1)
    plt.imshow(reconstructed_images[i], cmap=plt.cm.gray)
    plt.title("Reconstructed")
    plt.axis("off")

plt.tight_layout()
plt.show()
```
